```python
lst =  [[2, 6], [1, 2], [3, 4], [5, 3], [4, 1]]
```

```python
def sort(lst):
    for e in range(len(lst)):
        for i in range(len(lst)):
            if lst[i][0]>lst[e%len(lst)][0]:
                temp = lst[i]
                lst[i] = lst[e]
                lst[e] = temp
    return lst
```

```python
sort(lst)
```

```python
def divide_train_test(in_list, prop):
    row = int(len(in_list)*prop)
    train_data = in_list[:row]
    test_data = in_list[row:]
    return train_data, test_data
```

```python
print(divide_train_test(a_list ,0.7))
```

```python
print(divide_train_test(a_list, 0.9))
```

```python
def min_max_avg(lst):
    out = []
    out.append(min(lst))
    out.append(max(lst))
    out.append(sum(lst)/len(lst))
    return out
```

```python
def equ_interval(lst, interval):
    count = int(abs((lst[0]-lst[1])/interval))
    end = lst[0]
    out = []
    for i in range(interval):
        nend = end+count
        out.append([end, nend])
        end = nend
    return out
        
```

```python
equ_interval([-4,8],3)
```

```python
equ_interval([-2,10],4)
```

```python
equ_interval([3,18],5)
```
